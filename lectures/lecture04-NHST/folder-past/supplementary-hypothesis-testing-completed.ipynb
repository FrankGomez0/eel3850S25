Let's consider two coins: 

- coin1 has a probability of head being $p_1$

- coin2 has a probability of head being $p_2$.


we collect two sets of outcomes from coin1 and coin2, and aims to determine if $p_1=p_2$.

However, we do not know the exact value of $p_1$ or $p_2$.

First, we write a small program to collect data from these two coins:

```python
p1= 0.5 # assume a fair coin
p2= 0.1 # a biased coin

import math
import numpy as np
import numpy.random as npr
```

```python
samples1 = npr.choice([1,0], 100 , p= [p1,1-p1]) # flip coin1
samples2 = npr.choice([1,0], 100 , p= [p2,1-p2]) # flip coin2  
#print("the sample from coin1 is:", samples1)
#print("the sample from coin2 is:", samples2)
```

observations?

Let's construct a test statistics

```python
diff = np.abs(samples1.mean()- samples2.mean())
print("the absolute difference between the sample mean is:",  diff)
```

# Null Hypothesis testing

 - Null hypothesis: the two sample sets are generated by the same coin: $p_1=p_2$
 - Alternative Hypothesis: the two sample sets are generated by the different coin: $p_1\ne p_2$

# Let's assume the null hypothesis is true, what is the probability of observing the different this big? 

# Bootstrap sampling: Sampling with replacement.

```python
pooled = np.hstack((samples1,samples2))
pooled
```

```python
num_sims=10000
len1 = len(samples1)
len2 = len(samples2)
import random

count =0
for sim in range(num_sims):
    resamples1 = npr.choice(pooled,len1)
    resamples2 = npr.choice(pooled,len2)
    newdiff = np.abs( np.mean(resamples1)- np.mean(resamples2))
    if newdiff >= diff:
        count += 1
print("the probability of observing difference in two samples mean this big is ", count/num_sims)

```

# Conclusions:

$p$-value: measures the probability of obtaining the observed results, assuming that the null hypothesis is true.
    
- Because the $p$-value is greater than 0.05 in the null hypothesis, we cannot reject the null hypothesis.
